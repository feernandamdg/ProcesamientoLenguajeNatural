b) Cambie el tokenizador para que permita dobles o (más) espacios y/o caracteres especiales (signos de puntuación) entre palabras y aún así tokenice.
Ejem. Marco@Antonio___ Ortiz_/ --> Marco, Antonio, Ortiz.

```python
import string
import time
import tracemalloc

def tokenizador6(texto):
  t_ini = time.time()
  tracemalloc.start()

  texto_limpio = ""
  letras_permitidas = string.ascii_letters + "ñÑáéíóúÁÉÍÓÚ" + " "

  for caracter in texto:
    if caracter in letras_permitidas:
      texto_limpio += caracter

  print(f"Texto después de limpiar: '{texto_limpio}'")

  # considerando mayúsculas y espacios
  token = ""
  tokens = []

  for i in range(len(texto_limpio)):
    caracter = texto_limpio[i]

    # Si encontramos un espacio separamos
    if caracter == ' ':
      if token != "":
        tokens = tokens + [token]
        token = ""
    # Si encontramos una mayúscula (y no es el primer carácter), separamos
    elif caracter.isupper() and token != "" and i > 0 and texto_limpio[i-1] != ' ':
      tokens = tokens + [token]
      token = caracter
    else:
      token += caracter

  # Agregar el último token si existe
  if token != "":
    tokens = tokens + [token]

  t_fin = time.time()
  actual, pico = tracemalloc.get_traced_memory()
  tracemalloc.stop()

  print(f"El tiempo de ejecucion es de: {t_fin - t_ini} segundos")
  print(f"La memoria actual es de: {actual/10*3} Kb y la memoria pico fue de {pico/10*3} Kb")

  return tokens

# Pruebas
print("=" * 50)
print("PRUEBA 1: Ejemplo del profesor")
texto1 = "Marco@Antonio__ Ortiz/"
print("Texto original:", texto1)
print("Tokens finales:", tokenizador6(texto1))

print("\n" + "=" * 50)
print("PRUEBA 2: Texto con caso complejo")
texto2 = "Corn@@@e@jo__ Perez___Luna 302"
print("Texto original:", texto2)
print("Tokens finales:", tokenizador6(texto2))

print("\n" + "=" * 50)
print("PRUEBA 3: Texto original del ejercicio")
texto3 = "Los vie!@@rnes@ si   hay @ clase. P1ero sólo 302 minutos..."
print("Texto original:", texto3)
print("Tokens finales:", tokenizador6(texto3))
```

c) Modifique el tokenizador para que si hay números o carácteres en medio de la palabra, la tokenice. Pero si hay números a parte si los tokenice: Ejem. Mar12c@o 3312 --> Marco, 3312.

```python
import string
import time
import tracemalloc

def tokenizador5(texto):
  t_ini = time.time()
  tracemalloc.start()

  token = ""
  tokens = []
  delimitadores = string.whitespace + string.punctuation

  if texto[-1] not in delimitadores:
    texto = texto + '.'
    print("Texto ingresado:", texto)
  else:
    print("No se agregó punto:", texto)

  is_number = None

  for char in texto:
    if char == ' ' or char == '.':
      if token != "":
        tokens.append(token)
      token = ""
      is_number = None

    elif token == "" and char.isdigit(): #Si inicio de token es digito
        is_number = True
        token += char
    elif token == "" and char.isalpha(): #Si inicio de token es letra
        is_number = False
        token += char
    elif char.isalpha() and is_number: # Si comenzó como digito y encontró letra.
          token = char
          is_number = False
    elif (char.isdigit() and is_number) or (char.isalpha() and not is_number): #Si char corresponde a la bandera
          token += char

  t_fin = time.time()
  actual, pico = tracemalloc.get_traced_memory()
  tracemalloc.stop()
  print(f"El tiempo de ejecucion es de: {t_fin - t_ini} segundos")
  print(f"La memoria actual es de: {actual/10**3} Kb y la memoria pico fue de {pico/10**3} Kb")

  return tokens

texto = " Lo39090909090904s vi/////ernes si ///hay clase. 88Pero sólo 300 minutos..."
print(tokenizador5(texto))
```
